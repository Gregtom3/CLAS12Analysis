```python
import ROOT
c = ROOT.TCanvas("c","c",800,600)
ROOT.gStyle.SetOptStat(0)
ROOT.gStyle.SetPadLeftMargin(0.15)
ROOT.gStyle.SetPadRightMargin(0.25)
latex = ROOT.TLatex()
```

```python
file = ROOT.TFile("/work/clas12/users/gmat/CLAS12Analysis/macros/dihadron_analysis/rg-c/temp_data/rgc-test-eX.root","READ")
tree = file.Get("tree_reco")
```

```python
h_electron_2d = ROOT.TH2F("h_electron_2d","Electron Vertex (Wide view);vx [cm];vy [cm]",1000,-30,30,1000,-30,30)
tree.Draw("vy:vx>>h_electron_2d","pid==11 && vz!=-3","colz")
latex.DrawLatexNDC(0.5,0.83,"{:0.0f} events".format(h_electron_2d.GetEntries()))
c.Draw()
```

```python
h_electron_2d = ROOT.TH2F("h_electron_2d","Electron Vertex (Zoomed View);vx [cm];vy [cm]",1000,-10,10,1000,-10,10)
tree.Draw("vy:vx>>h_electron_2d","pid==11 && vx>-10&&vx<10&&vy>-10&&vy<10&& vz!=-3","colz")
latex.DrawLatexNDC(0.5,0.83,"{:0.0f} events".format(h_electron_2d.GetEntries()))
c.Draw()
```

```python
h_electron_1d = ROOT.TH1F("h_electron_1d","Electron Vz;vz [cm];Counts",250,-14,12)
tree.Draw("vz>>h_electron_1d","pid==11 && vz>-14&&vz<12&& vz!=-3","colz")
latex.DrawLatexNDC(0.5,0.83,"{:0.0f} events".format(h_electron_1d.GetEntries()))
c.Draw()
```
