```python
import ROOT
c = ROOT.TCanvas("c","c",800,600)
ROOT.gStyle.SetOptStat(0)
ROOT.gStyle.SetPadLeftMargin(0.15)
ROOT.gStyle.SetPadRightMargin(0.25)
latex = ROOT.TLatex()
```

```python
file = ROOT.TFile("/work/clas12/users/gmat/CLAS12Analysis/macros/dihadron_analysis/rg-c/temp_data/rgc-test-eX.root","READ")
tree = file.Get("tree_reco")
file_mc = ROOT.TFile("/volatile/clas12/users/gmat/clas12analysis.sidis.data/rgc-mc-plus-proton/aug16/merged_aug16.root","READ")
tree_mc = file_mc.Get("tree_MC")
```

```python
h_electron_E = ROOT.TH1F("h_electron_E",";E [GeV];Counts",100,0,10.6)
tree.Draw("E>>h_electron_E","pid==11 && status < 0","hist")
latex.DrawLatexNDC(0.3,0.8,"electron status < 0")
c.Draw()
```
