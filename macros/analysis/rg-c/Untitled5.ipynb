```python
import ROOT
import pandas as pd
import glob
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[]
y=[]
yerr=[]
for path in glob.glob("/volatile/clas12/users/gmat/clas12analysis.sidis.data/rgc-scaler/run*_HELScaler-all.csv"):
    df = pd.read_csv(path)
    run=df.run[0]
    npos=len(df[df.helicity==1])
    nneg=len(df[df.helicity==-1])
    x.append(run)
    y.append((npos-nneg)/(npos+nneg))
    yerr.append(2*np.sqrt((nneg**2*npos+npos**2*nneg)/((npos+nneg)**4)))
```

```python
plt.figure(dpi=150)
plt.errorbar(x,y,yerr=yerr,fmt=".")
plt.xlabel("Run")
plt.xlim(16600,17200)
plt.ylabel(r"Asymmetry in Helicity Counts $(N^{+}-N^{-})/(N^{+}+N^{-})$")
plt.grid()
```
