```python
def compare_run(run):
    rootfile = [path for path in Path(rootdir).glob("*/*{}*.root".format(run))][0]
    
    df = ROOT.RDataFrame("tree_postprocess",str(rootfile))
    
    gNp = df.Define("vz","abs(vz_e+4.5)").Filter("helicity==1 && p_e > 2.6 && th_e > 0.139626 && th_e < 0.610865 && vz < 4").Count().GetValue()
    gNm = df.Define("vz","abs(vz_e+4.5)").Filter("helicity==-1 && p_e > 2.6 && th_e > 0.139626 && th_e < 0.610865 && vz < 4").Count().GetValue()
    
    hNp = harutCSV[harutCSV[1]==run][2].to_numpy()[0]
    hNm = harutCSV[harutCSV[1]==run][3].to_numpy()[0]
    
    print("Run",run)
    print("Harut: N+ = {} , N- = {}".format(hNp,hNm))
    print("Gregory: N+ = {} , N- = {}".format(gNp,gNm))
```

```python
compare_run(16889)
```
